```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score
```

```python
# загрузка данных и разделение на обучающую и тестовую выборки
%matplotlib inline 
sns.set(style="ticks")
data = pd.read_csv('marvel-wikia-data.csv', sep=",", encoding='latin1')
data = data.dropna(axis=0, how='any')
data.drop(['name', 'urlslug', 'GSM', 'FIRST APPEARANCE'], axis=1, inplace=True)

data.head(10)
```

```python
data['ID'].unique()

ID_type_dict = {
    'Public Identity': 0,
    'Secret Identity': 1,
    'No Dual Identity': 2,
}

data['ID'] = data['ID'].map(ID_type_dict)
```

```python
data['ALIGN'].unique()

ALIGN_type_dict = {
    'Good Characters': 0,
    'Bad Characters': 1,
    'Neutral Characters': 2,
}

data['ALIGN'] = data['ALIGN'].map(ALIGN_type_dict)
```

```python
data['EYE'].unique()

EYE_type_dict = {
    'Blue Eyes': 0,
    'Brown Eyes': 1,
    'Green Eyes': 2,
    'Yellow Eyes': 3,
    'Red Eyes': 4,
    'Black Eyes': 5,
    'White Eyes': 6,
    'Grey Eyes': 7,
    'Hazel Eyes': 8,
    'Variable Eyes': 9,
}

data['EYE'] = data['EYE'].map(EYE_type_dict)
```

```python
data['HAIR'].unique()

HAIR_type_dict = {
    'Red Hair': 0,
    'Brown Hair': 1,
    'Purple Hair': 2,
    'Black Hair': 3,
    'Bald': 4,
    'Strawberry Blond Hair': 5,
    'Blond Hair': 6,
    'Silver Hair': 7,
    'White Hair': 8,
    'No Hair': 9,
    'Reddish Blond Hair': 10,
    'Variable Hair': 11,
}

data['HAIR'] = data['HAIR'].map(HAIR_type_dict)
```

```python
data['SEX'].unique()

SEX_type_dict = {
    'Male Characters': 0,
    'Female Characters': 1,
    'Genderfluid Characters': 2,
}

data['SEX'] = data['SEX'].map(SEX_type_dict)
```

```python
data['ALIVE'].unique()

ALIVE_type_dict = {
    'Living Characters': 0,
    'Deceased Characters': 1,
}

data['ALIVE'] = data['ALIVE'].map(ALIVE_type_dict)
data
```

```python
X = data.drop('SEX', axis=1)
y = data['SEX']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Стандартизация данных
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# ДЕРЕВО РЕШЕНИЙ
from sklearn.tree import DecisionTreeClassifier

# Создание и обучение модели дерева решений
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)

# Предсказание и оценка
y_pred_tree = tree_model.predict(X_test)
accuracy_tree = accuracy_score(y_test, y_pred_tree)
f1_tree = f1_score(y_test, y_pred_tree , average='weighted')

print(f'Decision Tree - Accuracy: {accuracy_tree}, F1-Score: {f1_tree}')
```

```python
# ГРАДИЕНТНЫЙ БУСТИНГ
from sklearn.ensemble import GradientBoostingClassifier

# Создание и обучение модели градиентного бустинга
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train, y_train)

# Предсказание и оценка
y_pred_gb = gb_model.predict(X_test)
accuracy_gb = accuracy_score(y_test, y_pred_gb)
f1_gb = f1_score(y_test, y_pred_gb, average='weighted')

print(f'Gradient Boosting - Accuracy: {accuracy_gb}, F1-Score: {f1_gb}')
```

```python
# РЕЗУЛЬТАТЫ
print(f'Decision Tree - Accuracy: {accuracy_tree}, F1-Score: {f1_tree}')
print(f'Gradient Boosting - Accuracy: {accuracy_gb}, F1-Score: {f1_gb}')
```

### <b>Выводы</b>
- Модель градиентного бустинга показала более низкие значения Accuracy и F1-Score по сравнению с моделью дерева решений, что указывает на худшую общую производительность.
- Модель дерева решений, хотя и является простой в интерпретации и реализации, но не уступает в качестве предсказаний более сложным моделям, таким как градиентный бустинг, особенно на комплексных данных.
