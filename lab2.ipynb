Обработка пропусков в данных, кодирование категориальных признаков, масштабирование данных.

Цель лабораторной работы: изучение способов предварительной обработки данных для дальнейшего формирования моделей.

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler, StandardScaler, Normalizer

# Загрузка данных
data = pd.read_csv('./coursera_courses.csv', sep=",")

# Проверка наличия пропущенных значений
print(data.isnull().sum())


```

```python
# Удаление строк с пропущенными значениями
data_cleaned = data.dropna(axis=0, how='any')

# Обработка числовых признаков
num_cols = data_cleaned.select_dtypes(include=['float64', 'int64']).columns

# Выполнение импьютации для числовых признаков
imputer = SimpleImputer(strategy='mean')
data_cleaned[num_cols] = imputer.fit_transform(data_cleaned[num_cols])

# Обработка категориальных признаков
cat_cols = data_cleaned.select_dtypes(include=['object']).columns

# Выполнение импьютации для категориальных признаков
for col in cat_cols:
    data_cleaned[col].fillna(data_cleaned[col].mode()[0], inplace=True)

# Кодирование категориальных признаков
label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    data_cleaned[col] = le.fit_transform(data_cleaned[col])
    label_encoders[col] = le

# Масштабирование данных
scaler = MinMaxScaler()
data_cleaned[num_cols] = scaler.fit_transform(data_cleaned[num_cols])

# Вывод обработанных данных
print(data_cleaned.head())
```
